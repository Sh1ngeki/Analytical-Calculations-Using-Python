**1. ᲰᲘᲞᲝᲗᲔᲖᲘᲡ ᲢᲔᲡᲢᲘᲠᲔᲑᲐ :**

_ინტერნეტ პროვაიდერი კომპანია ამტკიცებს, რომ მათი მომხმარებლების 60%-ზე მეტი იყენებს პრემიუმ პაკეტს. შერჩევაში 200 მომხმარებლიდან 105-ს აღმოაჩნდა პრემიუმ პაკეტი. შეამოწმეთ კომპანიის მტკიცება 0.05 მნიშვნელოვნების დონეზე (I ტიპის შეცდომის ალბათობის მიმღებლობის დონე)._


1.1. ნულოვანი ($H_0$) და ალტერნატიული ($H_1$) ჰიპოთეზების ჩამოყალიბება


```python


# მონაცემები
n = 200  # შერჩევის ზომა
x = 105  # წარმატებების რაოდენობა
pr_0 = 0.6   # საცდელი პროპორცია

print("ჰიპოთეზების ტესტირება პროპორციისთვის")
print("-" * 50)
# დაასრულეთ კოდი:
# 1. დაბეჭდეთ ნულოვანი ჰიპოთეზა
print("ნულოვანი ჰიპოთეზა (H0): მომხმარებლების 60%-ზე მეტი იყენებს პრემიუმ პაკეტს.")
# 2. დაბეჭდეთ ალტერნატიული ჰიპოთეზა
print("ალტერნატიული ჰიპოთეზა (H1): მომხმარებლების 60%-ზე ნაკლები იყენებს პრემიუმ პაკეტს.")
```

1.2. შერჩევის z-სტატისტიკის გამოთვლა


```python


import numpy as np

# მონაცემები
n = 200
x = 105
pr_0 = 0.6
pr_hat = x/n

# დაასრულეთ კოდი:
# 1. გამოთვალეთ შერჩევის სტანდარტული შეცდომა
std_error = np.sqrt(pr_0 * (1 - pr_0) / n)
print(f"შერჩევის სტანდარტული შეცდომა = {std_error:.3f}")

# 2. გამოთვალეთ z-სტატისტიკა
z_stat = (pr_hat - pr_0) / std_error
print(f"z-სტატისტიკა = {z_stat:.3f}")
```

1.3. p-მნიშვნელობის პოვნა და ამის მიხედვით გადაწყვეტილებების მიღება




```python


from scipy import stats

# მონაცემები
z_stat = -2.165
alpha = 0.05

# დაასრულეთ კოდი:
# 1. გამოთვალეთ p-მნიშვნელობა ცალმხრივი ტესტისთვის
p_value = 1 - stats.norm.cdf(z_stat)
print(f"p-მნიშვნელობა = {p_value:.4f}")


# 2. მიიღეთ გადაწყვეტილება
if p_value < alpha:
  print("მომხმარებლების 60%-ზე ნაკლები იყენებს პრემიუმ პაკეტს.")
else:
  print("მომხმარებლების 60%-ზე მეტი იყენებს პრემიუმ პაკეტს.")
```

1.4. ინტერპრეტაცია შედეგების

```python


# მონაცემები
n = 200
x = 105
pi_hat = x/n
p_value = 0.9848
alpha = 0.05

print("შედეგების ინტერპრეტაცია:")
print("-" * 50)
# დაასრულეთ კოდი:
# 1. დაბეჭდეთ შერჩევის პროპორცია პროცენტულად
print(f"შერჩევის პროპორცია: {pi_hat*100:.2f}%")

# 2. დაბეჭდეთ დასკვნა სტატისტიკური მნიშვნელოვნების შესახებ
if p_value < alpha:
    print("მომხმარებლების 60%-ზე ნაკლები იყენებს პრემიუმ პაკეტს.")
else:
  print("მომხმარებლების 60%-ზე მეტი იყენებს პრემიუმ პაკეტს.")
```

**2. ᲐᲛᲝᲕᲐᲠᲓᲜᲘᲚᲘ ᲛᲜᲘᲨᲕᲜᲔᲚᲝᲑᲔᲑᲘᲡ ᲞᲝᲕᲜᲐ  

2.1. ამოვარდნილი მნიშვნელობების დეპარტამენტის მიხედვით პოვნა, z ქულის (სტანდარტული გადახრის) მეთოდით და ვიზუალიზირება


```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# მონაცემების ჩატვირთვა
df = pd.read_csv('earnings.csv')

# z-ქულების გამოთვლა
df['z_score'] = (df['Salary'] - df['Salary'].mean()) / df['Salary'].std()

# ამოვარდნილი მნიშვნელობების ფილტრაცია z-ქულის მეთოდით
df_no_outliers_zscore = df[np.abs(df['z_score']) <= 3]


# box plot ვიზუალიზება
plt.figure(figsize=(10, 6))
df_no_outliers_zscore.boxplot(column='Salary', by='Department')

plt.title('ხელფასების განაწილება ამოვარდნილი მნიშვნელობების გარეშე (|z| > 3)')
plt.suptitle('იდკ')
plt.xlabel('დეპარტამენტი')
plt.ylabel('ხელფასი')
plt.show()

# შედეგების ბეჭდვა
print("\nამოვარდნილი მნიშვნელობების ანგარიში (z-score მეთოდი):")
print(f"სრული მონაცემები: {len(df)}")
print(f"ამოვარდნილი მნიშვნელობების გარეშე: {len(df_no_outliers_zscore)}")
print(f"ამოვარდნილი მნიშვნელობების რიცხვი: {len(df) - len(df_no_outliers_zscore)}")
```

2.2. ამოვარდნილი მნიშვნელობები დეპარტამენტების მიხედვით IQR (ინტერკვარტილური დიაპაზონის) მეთოდით პოვნა და ვიზუალიზება. 

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# მონაცემების ჩატვირთვა
df = pd.read_csv('earnings.csv')

# IQR-ის გამოთვლა
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

# ზღვრების განსაზღვრა
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


# ამოვარდნილი მნიშვნელობების ფილტრაცია IQR მეთოდით
df_no_outliers_iqr = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]


# ვიზუალიზება
plt.figure(figsize=(10, 6))
df_no_outliers_iqr.boxplot(column='Salary', by='Department')
plt.title('ხელფასების განაწილება ამოვარდნილი მნიშვნელობების გარეშე (IQR მეთოდი)')
plt.suptitle('')
plt.xlabel('დეპარტამენტი')
plt.ylabel('ხელფასი')
plt.show()

# შედეგების ბეჭდვა
print("\nამოვარდნილი მნიშვნელობების ანგარიში (IQR მეთოდი):")
print(f"სრული მონაცემები: {len(df)}")
print(f"ამოვარდნილი მნიშვნელობების გარეშე: {len(df_no_outliers_iqr)}")
print(f"ამოვარდნილი მნიშვნელობების რიცხვი: {len(df) - len(df_no_outliers_iqr)}")
```
